```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import dataloader
from torchvision import transforms
from torchvision import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
```

```python
transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(0.1307,0.3081)
])

train_data=datasets.MNIST(
    root='E:/Jupyter/dataset/mnist',
    train=True,
    download=True,
    transform=transform
)

test_data=datasets.MNIST(
    root='E:/Jupyter/dataset/mnist',
    train=False,
    download=True,
    transform=transform
)
```

```python
# 安全导入 DataLoader
def import_dataloader():
    """安全导入 DataLoader 函数"""
    try:
        from torch.utils.data import DataLoader
        return DataLoader
    except ImportError:
        from torch.utils.data.dataloader import DataLoader
        return DataLoader

# 使用
DataLoader = import_dataloader()

# 现在可以正常使用了
train_loader = DataLoader(train_data, shuffle=True, batch_size=256)
test_loader = DataLoader(test_data, shuffle=False, batch_size=256)
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1, 6, kernel_size=5, padding=2), nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),
            nn.Conv2d(6, 16, kernel_size=5), nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),
            nn.Conv2d(16, 120, kernel_size=5), nn.Tanh(),
            nn.Flatten(),
            nn.Linear(120, 84), nn.Tanh(),
            nn.Linear(84, 10)
        )

    def forward(self, x):
        y = self.net(x)
        return y
```

```python
x=torch.rand(size=(1,1,28,28))
for layer in CNN().net:
    x=layer(x)
    print(layer.__class__.__name__, 'output_shape:\t', x.shape)
```

```python
model = CNN()
```

```python
loss_fn=nn.CrossEntropyLoss()
```

```python
learning_rate = 0.9
optimizer = torch.optim.SGD(
    model.parameters(),
    lr = learning_rate,
)
```

```python
epochs = 5
losses=[]

for epoch in range(epochs):
    for(x,y) in train_loader:
        
        Pred=model(x)
        loss=loss_fn(Pred,y)
        losses.append(loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
Fig=plt.figure
plt.plot(range(len(losses)),losses)
plt.show
```

```python
correct = 0
total=0

with torch.no_grad():
    for(x,y)in test_loader:
        Pred = model(x)
        _,predicted=torch.max(Pred.data,dim=1)
        correct+=torch.sum((predicted==y))
        total+=y.size(0)
        
print(f'测试集精准度:{100*correct/total}%')
```
